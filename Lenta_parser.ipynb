```python
# Импорт библиотек
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup

import random

import pandas as pd

from tqdm import tqdm
```

```python
# Настройка количества нажатий кнопки "показать еще"
number_of_button_clicks = 100

# Открываем браузер
browser = Chrome()
print("Браузер открылся...")
# Пауза на 3 секунды, чтобы браузер прогрузился
time.sleep(3)

# Cоздаем переменную со ссылкой
url = "https://lenta.ru/"
# Выполняем переход на сайт
browser.get(url)
print("Переход на сайт выполнен...")
# Пауза 5 секунд, чтобы сайт прогрузился
time.sleep(5)
print("Сайт запустился...")

# Создаем объект с главной кнопкой, которая ищется в HTML коде страницы по XPATH
main_button = browser.find_element(By.XPATH, '//*[@id="body"]/div[3]/div[3]/div[2]/header/div[2]/button')
# Нажимает главную кнопку
main_button.click()
print("Main_button нажата...")
time.sleep(3)

# Создаем объект с полем ввода, которое ищется в HTML коде страницы по XPATH
input_field = browser.find_element(By.XPATH, '//*[@id="body"]/div[3]/div[2]/div/div[2]/nav/ul/li[26]/div/input')
# Заполняет поле input_field
input_field.send_keys('Сбербанк')
print("Поле input_field заполнено...")
time.sleep(1)

# Создаем объект с кнопкой поиска, которая ищется в HTML коде страницы по XPATH
search_button = browser.find_element(By.XPATH, '//*[@id="body"]/div[3]/div[2]/div/div[2]/nav/ul/li[26]/div/button')
# Нажимаем кнопку поиска
search_button.click()
print("Переход в архив...")
time.sleep(15)
print("Сбор данных...")

# Нажимаем кнопку "показать еще", чтобы показать больше новостей
more_data_button = browser.find_element(By.XPATH, '//*[@id="body"]/div[3]/div[3]/main/div[2]/section/div[1]/ul/li[11]/div/button')
# Нажимаем кнопку с помощью цикла. tqdm делает progress bar
for i in tqdm(range(number_of_button_clicks)):
    # Между нажатиями делается рандомная пауза от 1.9 до 3.5 секунд, чтобы прогружались страницы и не был замечен робот
    time.sleep(round(random.uniform(1.9, 3.5), 2))
    more_data_button.click()
time.sleep(1)

# Загружаем в soup код страницы
soup = BeautifulSoup(browser.page_source, 'lxml')
print('Собрано заголовков:', 
      len(soup.findAll('li', class_='search-results__item _news')))

# Создаем датафрейм, в которую будут записываться нужные данные с целой html страницы
df = pd.DataFrame(
    {
        'Name': [], # Заголовок новости
        'Date': [], # Дата выхода новости
        'Link': []  # Ссылка на новость
    }
)

# В all_news сохраняем только информацию из новостей, которая содержится в заголовках
# search-results__item _news
all_news = soup.findAll('li', class_='search-results__item _news')
# Перебираем все собранные новости и разбиваем на нужные подразделы
for news in tqdm(all_news):
    name = news.find('h3', class_='card-full-news__title').text.strip()
    date = news.find('time', class_='card-full-news__info-item').text.strip()
    link = news.find('a', class_='card-full-news _search').get('href')
    link = 'https://lenta.ru/' + link
    # Добавляем в датафрейм полученную информацию в соотв. столбцы
    df.loc[(len(df.index))] = [name, date, link]
    time.sleep(0.01)
print("Данные загружены в DataFrame...")
time.sleep(1)
print("Работа закончена!")
df
```
