```python
# Импорт библиотек
import pandas as pd
import re
from datetime import timedelta
from datetime import datetime
```

```python
# Загружаем файл с 4426 результатов
data = pd.read_csv('df_4426_news.csv')
del data['Unnamed: 0']
```

```python
data.head(1)
```

```python
# Словарь для перевода месяца в нужный результат
dict_month = {'января': '01', 
              'февраля': '02', 
              'марта': '03',
              'апреля': '04', 
              'мая': '05', 
              'июня': '06',
              'июля': '07', 
              'августа': '08', 
              'сентября': '09',
              'октября': '10', 
              'ноября': '11', 
              'декабря': '12'}

# Словарь для перевода числа в нужный формат 
dict_days = {'1': '01',
             '2': '02',
             '3': '03',
             '4': '04',
             '5': '05',
             '6': '06',
             '7': '07',
             '8': '08',
             '9': '09'}

# Функция конвертации 
def date_converter(row):
    date = row[1]
    date = date.split()
    new_date = []
    new_date.append(date[3])
    new_date.append(date[2])
    new_date.append(date[1])
    for key, name in dict_month.items():
        if new_date[1] == key:
            new_date[1] = name
    for key, name in dict_days.items():
        if new_date[2] == key:
            new_date[2] = name
    return '-'.join(new_date)
```

```python
# Применяю функцию date_converter и конвертирую дату в нужный формат
data['Converted_data'] = data.apply(date_converter, axis=1)

# Группирую по дате, чтобы узнать количество заголовков на каждый день и сортирую по убыванию даты
grouped_data = data.groupby('Converted_data').count().sort_values('Converted_data', ascending=False).reset_index()
grouped_data.head()
```

```python
# Убираю ненужыне столбцы и даю новые имена двум оставшимся
grouped_data = grouped_data[['Converted_data', 'Name']]
grouped_data.columns = ['Date', 'Count']
grouped_data.head(5)
```

```python
# Создаю датафрейм для всех дат
df_dates = pd.DataFrame(
    {
        'Date': [],
    }
)
```

```python
# Заполняю датафрейм всеми датами подряд начиная с 15 января 2015 года
start_date = '2023-12-12'
end_date = '2015-01-01'
print(type(start_date))
start_date_datetime = datetime.strptime(start_date, '%Y-%m-%d')
end_date_datetime = datetime.strptime(end_date, '%Y-%m-%d')
print(start_date_datetime)
while start_date_datetime >= end_date_datetime:
    df_dates.loc[(len(df_dates.index))] = re.findall('\d{4}-\d{2}-\d{2}', str(start_date_datetime))
    start_date_datetime = start_date_datetime - timedelta(days = 1)
```

```python
df_dates.head(10)
```

```python
grouped_data.head(5)
```

```python
# Записываю даты новостей в список
grouped_data_dates_list = grouped_data['Date'].to_list()
grouped_data_dates_list
```

```python
# Функция для проверки, есть ли хотя бы одна новость в каждую дату. Если в какую-то
# из дат не было новостей, то этой даты вообще нет в датафрейме с новостями, туда
# добавится строка с недостающей датой и количество новостей 0 (так как их нет)
def compare(row):
    if row[0] not in grouped_data_dates_list:
        grouped_data.loc[(len(grouped_data.index))] = [row[0], 0]
```

```python
# Результат по новостному блоку. 
# Применяю функцию и сортирую по убыванию
df_dates.apply(compare, axis=1)
grouped_data = grouped_data.sort_values('Date', ascending=0).reset_index()
del grouped_data['index']
grouped_data.head(10)
```

```python
# Записываю получившийся датафрейм в CSV для передачи в работу коллегам
grouped_data.to_csv('NEWS.csv', index=False)
```

```python
price_data = pd.read_csv('SBER_01122010_12122023.csv')
del price_data['<DATE>']
price_data = price_data.reset_index()
price_data.columns = ['Date', 'ThirstPrice', 'High', 'Low', 'LastPrice', 'oper']
del price_data['oper']
price_data['Volatility RUB'] = price_data['High'] - price_data['Low']
price_data['Volatility %'] = ((price_data['High'] - price_data['Low'])\
/ ((price_data['LastPrice'] + price_data['ThirstPrice']) / 2)) * 100
price_data = price_data.sort_values('Date', ascending=False).reset_index()
del price_data['index']
price_data
```

```python
def convert_price_data(row):
    data = str(row[0])
    data = data[0] + data[1] + data[2] + data[3] + '-' + data[4]\
    + data[5] + '-' + data[6] + data[7]
    return data
```

```python
price_data['Date'] = price_data.apply(convert_price_data, axis=1)
```

```python
price_data.head(2)
```

```python
price_data_list = price_data['Date'].to_list()
```

```python
def compare_price_data(row):
    if row[0] not in price_data_list:
        #print(type(row[0]))
        price_data.loc[(len(price_data.index))] = [row[0], 0, 0, 0, 0, 0, 0]
```

```python
df_dates.apply(compare_price_data, axis=1)
price_data = price_data.sort_values('Date', ascending=0)
price_data = price_data.reset_index()
del price_data['index']
```
